```python
import matplotlib.pyplot as plt

plt.plot([1,2,9,4,3],[4,2,8,9,2])
plt.show()
```

```python
# multiline
import numpy as np
x=np.linspace(0,5,10)
y=np.linspace(2,6,10)
plt.xlabel("x label")
plt.ylabel("y Label")
plt.title("Multi line graph",color="blue")
plt.plot(x,y,"r^",x*2,y**3,"b-.",x/2,y*79,"y2")
plt.grid()
plt.show()
```

```python
plt.plot([8,3,9],[1,8,3],label="label1")
plt.plot([6,8,88],[3,2,6],label="label2")
plt.legend()
plt.grid()
plt.title("Multiplot")
plt.xlabel("x label")
plt.ylabel("ylabel")
plt.show()
```

```python
# subplots
x=np.linspace(1,10,100)
y=np.log(x)

plt.figure(1)

plt.subplot(2,2,1)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("y=log(x)**2")
plt.plot(x,y**2)

plt.figure(2)

plt.subplot(2,2,1)
plt.title("logx")
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title("Exponetial")
plt.plot(x,x**2)

plt.show()
```

```python
from matplotlib import style
x=np.linspace(1,38,53)
y=np.linspace(5,78,53)
style.use("ggplot")
plt.plot(x**2,y,label="xx",linewidth=10)
plt.title("style",color="blue")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.grid(True,color="b")
plt.show()
```

```python
plt.plot()
plt.bar([2,56,78,98,45,34],[56,45,76,87,34,45],label="kk",width=0.9)
plt.bar([87,67,54,90,34,76],[98,76,56,45,89,55],label="rk",width=1.7,color="y")
plt.legend()
plt.show()
```

```python
# PieChart
per=[45,67,34,78]
colors=["red","green","blue","yellow"]
explode=[0.1,0.01,0.01,0.01]
label=["c","java","c++","python"]
plt.pie(per,labels=label,explode=explode,colors=colors)
plt.show()
```
